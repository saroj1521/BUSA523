```python
#import random library
import random

#creat a variable that stores Head and Tail 
cointoss =["Head", "Tail"]

#create a new list to store all our cointoss value 
newlist=[]

#loop 10000 and store all the result in newlist
for tossoutput in range(10000):
    newlist.append(random.choice(cointoss))

print(newlist)

#lets assume totalstreaks is 0
totalstreaks=0

#lets assume streak is 0 as well
streak=0

#loop to all 10000 toss and check if there is any streaks of 5 or more
for tossoutput in range(len(newlist)):
    #if the next toss is same as previous toss then add to the value of streak
    if newlist[tossoutput]==newlist[tossoutput-1]:
        streak=streak+1
        
    #if not then there is no streak
    else:
        streak=0
            
    #when the streak is 5, add 1 to the totalstreak
    if streak ==5:
        totalstreaks=totalstreaks+1
        
print("The number of streaks from this experiment is " + str(totalstreaks))
```
