### Answer Q2:

```python
#This is the list of numbers given in the question
givenlist = [-6, 5, -3, -2, 1, 0, -8, 9, 3]

#define a function that that seperates positive and negative numbers
def positive_and_negative(anylist):
    #create two separate lists to seperate positive and negative numbers 
    positive=[]
    negative=[] 
    
    #loop that checks the number and add them to the above lists
    for value in anylist:
        if value>0:
            positive.append(value)
            
        else:
            negative.append(value)
    #return the negative and positive value
    return negative, positive

# Call the function and print it
print("Output: ", positive_and_negative(givenlist))
```

### Answer Q3:

```python
#define a function for winning possibilities of any final score
def winpossib(finalscore):
    #Put all the winnning possibilities in this list
    output = []
    #for any first card from 1 to 11
    for firstcard in range(1, 12):
        #the second card is between whatever the first card and 11
        for secondcard in range(firstcard, 12):
            #set the condition that has input final score always higher but less than 21
            if(finalscore < firstcard + secondcard and firstcard + secondcard <22):
                #add it to the list
                output.append([firstcard, secondcard])
    print("All Winning Possibilites are: ", output)

#create an input for the player's final score
playerscore = int(input("Enter the player's final score: "))

winpossib(playerscore)
```
